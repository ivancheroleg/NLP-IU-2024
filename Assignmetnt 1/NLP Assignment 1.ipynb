## Task 2. Implement [Byte-Pair Encoding(BPE)](https://arxiv.org/pdf/1508.07909.pdf) Tokenizer (80 points)

### Task 2.1. Implementation (60 points)

Implement the tokenizer as the BPETokenizer class:
* Implement `train` method that learns merges and builds the vocabulary of the specified `vocab_size` (25 points).
* Implement `tokenize` method that should tokenize the text according to the learnt merges (25 points).

Your code should have docstrings and comments (10 points).

```python
from typing import List
from collections import defaultdict
import re

class BPETokenizer:
    """
    Byte-Pair Encoding Tokenizer
    """

    def __init__(self, vocab_size: int) -> None:
        """
        :param vocab_size: the size of the vocabulary
        """
        self.vocab_size = vocab_size
        self.merges = {}
        self.word_frequencies = defaultdict(int)
        self.regex = r'@\w+|#\w+|http://\S+|\w+|:-[)]|[^a-zA-Z0-9\s]'

    def train(self, corpus: List[str]) -> None:
        """
        Learns merges and builds the vocabulary of the specified `vocab_size`
        :param corpus: list of strings, where each string is a sentence
        :return: None
        """
        
        # corpus is a list of strings, where each string is a sentence
        for sentence in corpus:
            # split the sentence into words (links, @usernames, ':-)' and #hashtags are treated as words)
            words = re.findall(self.regex, sentence.lower())
            for word in words:
                self.word_frequencies[word] += 1
        
        # build the vocabulary for letters
        letters = set("".join(self.word_frequencies.keys()))
        vocab = list(letters)
        vocab.sort()
        
        # if the vocab_size is not specified, set it to the size of the vocabulary initially
        if self.vocab_size == 0: self.vocab_size = len(vocab)
        
        # split the word into characters so then we can merge them
        splitted_words = {}
        for word in self.word_frequencies:
            splitted_words[word] = [l for l in word]

        # main loop for learning merges
        while len(vocab) < self.vocab_size:
            # count the frequency of pairs of characters
            pairs = defaultdict(int)
            
            for word, freq in self.word_frequencies.items():
                chars = splitted_words[word]
                # count the frequency of pairs of characters
                for i in range(len(chars) - 1):
                    pairs[chars[i], chars[i+1]] += freq
            
            # find the most frequent pair of characters
            max_freq_pair = max(pairs, key=pairs.get)
            
            # merge the most frequent pair of characters
            for word in self.word_frequencies:
                chars = splitted_words[word]
                
                # if the word is shorter than 2 characters, skip it
                if len(chars) < 2: continue
                
                # merge the most frequent pair of characters
                i = 0
                while i < len(chars) - 1:
                    if chars[i+1] == max_freq_pair[1] and chars[i] == max_freq_pair[0]:
                        chars = chars[:i] + [max_freq_pair[0] + max_freq_pair[1]] + chars[i+2:]
                    else:
                        i += 1
                splitted_words[word] = chars
            
            # update the vocabulary
            vocab.append(max_freq_pair[0] + max_freq_pair[1])
            self.merges[max_freq_pair] = max_freq_pair[0] + max_freq_pair[1]
            
    def tokenize(self, text: str) -> List[str]:
        """
        Tokenizes the text according to the learnt merges
        :param text: the input text
        :return: the list of tokens
        """
        # split the text into words (links, @usernames, ':-)' and #hashtags are treated as words)
        words = re.findall(self.regex, text.lower())
        
        splitted_words = []
        for word in words:
            splitted_words.append([l for l in word])
        
        # merge the characters according to the learnt merges
        for pair, merge in self.merges.items():
            for j, split in enumerate(splitted_words):
                i = 0
                while i < len(split) - 1:
                    if split[i] == pair[0] and split[i + 1] == pair[1]:
                        split = split[:i] + [merge] + split[i + 2:]
                    else:
                        i += 1
                splitted_words[j] = split
            
        return sum(splitted_words, []) 
```

### Task 2.2. Analysis on Tweets Dataset (10 points)

Train the BPE tokenizer on the tweets dataset. Try to tokenize the tweets with the tokenizer of different `vocab_size`. For example, train the BPE tokenizer with `vocab_size` of [base_vocab_size, 250, 500, 750, 1000]. Plot the dependency of the average length of the tokenized tweet by `vocab_size` to analyze how `vocab_size` affects the length of the tokenized tweet on average. Tell what `vocab_size` is preferrable and why.

```python
files = ['file1', 'file2', 'file3', 'file4', 'file5']
tweets = []
for file in files:
    with open('Assignment1_data/' + file, 'r',encoding="utf8") as f:
        tweets.append(f.read().split("\n"))

tweets = sum(tweets, [])
tweets = [tw for tw in tweets if tw != '']

print("Tweets at all:", len(tweets))
```

```python
vocab_sizes = [0, 250, 500, 750, 1000]
tweet_avg_lengths = []

for vocab_size in vocab_sizes:
    tokenizer = BPETokenizer(vocab_size)
    tokenizer.train(tweets)
    
    tokenized_tweets = [tokenizer.tokenize(tweet) for tweet in tweets]
    tweet_avg_lengths.append(sum([len(tw) for tw in tokenized_tweets]) / len(tokenized_tweets))
    print("With vocab size:", vocab_size, ", We have", tweet_avg_lengths[-1], "- average length of tokenized tweet")
```

```python
import matplotlib.pyplot as plt

plt.plot(vocab_sizes, tweet_avg_lengths)
plt.xlabel('Vocab size')
plt.ylabel('Average length of tokenized tweet')
plt.title('Dependency of the average length of the tokenized tweet by vocab size')
plt.show()
```

The trend shows with the increase of the vocab size, the average length of the tokenized tweet decreases. It is because the more vocab size we have, the more words we can represent with a single token. 
So, the preferable vocab size is 1000, because it gives the shortest average length of the tokenized tweet.
